```python
list_sum_Titian_Boday_index=list(sum_Titian_Boday.index)
#print(list_sum_Titian_Boday_index)

titian_need_Zero=[]
for i in range(len(zero_body[0])):
    titian_need_Zero.append(list_sum_Titian_Boday_index[zero_body[0][i]])
#print(titian_need_Zero)

def wrong_function(row):
    #print(row['泰坦名称'])
    #print("===========================")
    if str(row['泰坦名称']) in titian_need_Zero:
        #print("我进if啦==================================================")
        num_titian=list_sum_Titian_Boday_index.index(row['泰坦名称'])
        #print(f"我是第{num_titian}个泰坦")
        wrong_dmg=0
        #遍历zero_body[0]，如果和num_titian一样，就累加，遍历结束，返回
        for i in range(len(zero_body[0])):
            if num_titian==i:
                wrong_dmg += row[armor[zero_body[1][i]]]
                #print(wrong_dmg)
        return wrong_dmg
    else:
        return 0



df["错误伤害"]=df.apply(wrong_function,axis=1,args=())

df["有效伤害"]=df["泰坦伤害"]-df["错误伤害"]

df_total=df.groupby(["玩家编号","玩家名称"]).sum()
df_total["突袭次数"]=df.groupby(["玩家编号","玩家名称"])["总突击伤害"].mean()
df_total["个人均伤"]=df_total["泰坦伤害"]/df_total["突袭次数"]
df_total["有效占比"]=round(df_total["有效伤害"]/df_total["泰坦伤害"]*100,2)

df_total['armor_total'] = round(df_total[armor].apply(lambda x : (x.sum()),axis=1 ),2)

df_total['body_total'] = round(df_total[body].apply(lambda x : (x.sum()),axis=1 ),2)

df_total['armor_total_per']=round((df_total['armor_total']/df_total["泰坦伤害"])*100,2)

df_total['body_total_per']=round((df_total['body_total']/df_total["泰坦伤害"])*100,2)

df_total["Rank_有效伤害"]=df_total["有效伤害"].rank(ascending=False).map(lambda x:int(x))
df_total["Rank_倒数次数"]=df_total["总突击伤害"].rank(ascending=True).map(lambda x:int(x))

df_f=df_total.drop(columns=['总突击伤害',"泰坦编号",'头部(装甲)', '身体(装甲)','左臂(装甲)', '右臂(装甲)', '左手(装甲)', '右手(装甲)', '左腿(装甲)', '右腿(装甲)',
                       '头部(肉体)','身体(肉体)', '左臂(肉体)', '右臂(肉体)', '左手(肉体)', '右手(肉体)', '左腿(肉体)', '右腿(肉体)',
                       '头部(骨架)', '身体(骨架)', '左臂(骨架)', '右臂(骨架)', '左手(骨架)', '右手(骨架)', '左腿(骨架)','右腿(骨架)' ])


list_columns_temp=list(df_f.columns)

timestamp=pd.Timestamp(file_year+"-"+file_month+"-"+file_day)

df_f.columns=[[timestamp]*len(list(df_f.columns)),list_columns_temp]


dataframe_lists.append(df_f)
```
