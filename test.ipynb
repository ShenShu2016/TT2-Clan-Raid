```python
for file in filename_list:
    if file[-3:]=="csv":
        file_year=file[:4]  #STR
        file_month=file[4:6]
        file_day=file[6:8]
        print("Timestamp="+file_year+"-"+file_month+"-"+file_day)


        df = pd.read_csv(filePath+file,encoding='utf-8')

        armor = ['头部(装甲)', '身体(装甲)','左臂(装甲)', '右臂(装甲)', '左手(装甲)', '右手(装甲)', '左腿(装甲)', '右腿(装甲)']
        body = ['头部(肉体)','身体(肉体)', '左臂(肉体)', '右臂(肉体)', '左手(肉体)', '右手(肉体)', '左腿(肉体)', '右腿(肉体)']
        skeleton = ['头部(骨架)', '身体(骨架)', '左臂(骨架)', '右臂(骨架)', '左手(骨架)', '右手(骨架)', '左腿(骨架)','右腿(骨架)']

        sum_Titian_Boday=df.groupby("泰坦名称").sum()[body]

        grouped=np.where(sum_Titian_Boday==0)
        zero_body=np.array(grouped).tolist()
        #print(zero_body)#first list is for titians, second list is for Armor

        for i in range(len(zero_body[0])):
            print(f"{file}\t{sum_Titian_Boday.index[zero_body[0][i]]}\t不打\t{armor[zero_body[1][i]]}")
```
